```python
import folium
m=folium.Map(location=[6.45420, 3.38934])
m
```

```python
from branca.element import Figure
fig=Figure(width=550,height=350)
```

```python
m1=folium.Map(width=550,height=350,location=[6.45420, 3.38934],zoom_start=11,min_zoom=8,max_zoom=16)
fig.add_child(m1)
m1
```

```python
fig2=Figure(width=550,height=350)
m2=folium.Map(location=[6.45420, 3.38934])
fig2.add_child(m2)
folium.TileLayer('Stamen Terrain').add_to(m2)
folium.TileLayer('Stamen Toner').add_to(m2)
folium.TileLayer('Stamen Water Color').add_to(m2)
folium.TileLayer('Cartodbpositron').add_to(m2)
folium.TileLayer('Cartodbdark_matter').add_to(m2)
folium.LayerControl().add_to(m2)
m2
```

```python
# Create Basemap
fig3=Figure(width=550,height=350)
m3=folium.Map(location=[6.45420, 3.1004],tiles='cartodbpositron',zoom_start=15)
fig3.add_child(m3)

#Add markers to basemap
folium.Marker(location=[6.45420, 3.38943],popup='Default popup Marker1',tooltip='Click here to view Popup').add_to(m3)
folium.Marker(location=[6.53122, 3.61404],popup='<strong>Marker2</strong>',tooltip='<strong>Click here to view Popup</strong>').add_to(m3)
folium.Marker(location=[6.75443, 3.1004],popup='<h3 style="color:green;">Marker3</h3>',tooltip='<strong>Click here to view Popup</strong>').add_to(m3)

m3
```
