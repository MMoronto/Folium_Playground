```python
df_covid=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Nigeria_covid_demo.csv')
df_covid.head()
```

```python
id_dict={'Abia': '0',
  'Abuja': '1',
  'Adamawa': '2',
  'Akwa Ibom': '3',
  'Anambra': '4',
  'Bauchi': '5',
  'Benue': '6',
  'Bornu': '7',
  'Cross River': '8',
  'Delta': '9',
  'Edo': '10',
  'Ebonyi': '11',
  'Ekiti': '12',
  'Enugu': '13',
  'Jigawa': '14',
  'Imo': '15',
  'Kaduna': '16',
  'Kano': '17',
  'Katsina': '18',
  'Kogi': '19',
  'Kwara': '20',
  'Lagos': '21',
  'Nasarawa': '22',
  'Niger': '23',
  'Ogun': '24',
  'Ondo': '25',
  'Osun': '26',
  'Oyo': '27',
  'Plateau': '28',
  'Rivers': '29',
  'Sokoto': '30',
  'Taraba': '31',
  'Yobe': '32'}

# Create IDs using the Dictionary above
df_covid['state_id']=df_covid['Name of State'].map(id_dict)
```

```python
df_covid['Active Cases']=df_covid['Total Confirmed cases']-(df_covid['Cured/Discharged/Migrated']+df_covid['Death'])
df_covid.head()
```

```python
bins=np.linspace(min(df_covid['Active Cases']),max(df_covid['Active Cases']),11)
bins
```

```python
# Color states with active COVID-19 cases
df_covid['color']=pd.cut(df_covid['Active Cases'],bins,labels=['#FFEBEB','#F8D2D4','#F2B9BE','#EBA1A8','#E58892','#DE6F7C','#D85766','#D13E50','#CB253A','#C50D24'],include_lowest=False)

#Color states with no active cases previously had
df_covid['color'].replace(np.nan, '#32CD32',inplace=True)
```

```python
df_covid=df_covid[['Date','state_id','color']]
df_covid.head()
```

```python
df_covid['Date']=(df_covid['Date'].astype(int)// 10**9).astype('U10')
covid_dict={}
for i in df_covid['state_id'].unique():
    covid_dict[i]={}
    for j in df_covid[df_covid['state_id']==i].set_index(['state_id']).values:   
        covid_dict[i][j[0]]={'color':j[1],'opacity':0.7}
```
